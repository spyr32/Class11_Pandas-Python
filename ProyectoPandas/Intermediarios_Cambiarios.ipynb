```python
import pandas as pd

#Se realiza el proceso de web scrapping
df_web = pd.read_html('https://gee.bccr.fi.cr/IndicadoresEconomicos/Cuadros/frmConsultaTCVentanilla.aspx',encoding='utf-8', decimal=',', thousands='.')
```

```python
df_web[2]
```

## Trabajamos en limpiar los datos de la tabla importada

```python
#Importe 38 registro y 6 columnas, debe eliminar dos registros mal importados
print(df_web[2].shape)

#Guardo el dataframe en una variable nueva en memoria
df = df_web[2].copy()
del(df_web) #Elimino el df importado
```

```python
#Imprime encabezado de df
df.head(3)
```

```python
#Imprime los titulos de la columnas
df.columns
```

### Paso1: Lipiar nombres de columnas

- Extraer la fila 0
- Asigno fila 0 como columna

```python
#Extraer fila 0
primera_fila = df.iloc[0]

#Asignar las etiquetas a las columnas
df.columns = primera_fila

#Borrar Fila 0
df.drop(0, inplace=True)

df.head()
```

```python
#Renombrar las columnas
df.rename(columns={'Tipo de Entidad':'Tipo','Entidad Autorizada':'Entidad','Diferencial Cambiario':'Diferencial'}, inplace=True)

df.columns
```

```python
#Eliminar series con algún valor nulo
df.dropna()
```

```python
#Completa valores nulos con un valor en especifico
df['Tipo'].fillna('NO valor').tail()
```

```python
# Remplazar valores NaN con el último valor no nulo en la misma columna
df['Tipo'].ffill(inplace=True)
```

```python
df
```

```python
#Opción1 eliminar con index
df.drop(df.tail(1).index, inplace= True)

#Opción2: eliminar con dropna
df.dropna(inplace=True)
```

# Realizar análisis sobre los datos limpios

```python
# Consultar la entidad con la venta mas baja
tipo_venta_min = df['Venta'].min()


entidades_con_min = df[df['Venta'] == tipo_venta_min]['Entidad']

print('Tipo cambio minimo: ', tipo_venta_min)
print('Entidad/s', entidades_con_min)
```

```python
#Conultar la entidad con venta mal alta
tipo_venta_max = df['Venta'].max()

entidades_con_max = df[df['Venta'] == tipo_venta_max]

entidades_con_max
```

```python
entidades_con_max = df[df['Venta'] == tipo_venta_max]['Entidad']
entidades_con_max
```
